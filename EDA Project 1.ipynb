```python
import pandas as pd
import numpy as np

df = pd.read_csv('iris.data', header=None)
df
```

```python
#Histogram of each attribute

import matplotlib.pyplot as plt
plt.hist(df[0], bins=10, color = 'red') 
plt.title(f'Histogram for sepal length in cm')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()

plt.hist(df[1], bins=10,) 
plt.title(f'Histogram for sepal width in cm')
plt.xlabel('Width')
plt.ylabel('Frequency')
plt.show()

plt.hist(df[3], bins=10, color = 'green')  
plt.title(f'Histogram for petal length in cm')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()

plt.hist(df[3], bins=10, color = 'yellow')  
plt.title(f'Histogram for petal width in cm')
plt.xlabel('Width')
plt.ylabel('Frequency')
plt.show()

df['CategoryCode'] = pd.Categorical(df[4]).codes
plt.hist(df['CategoryCode'], bins=len(df[4].unique()))
plt.xticks(range(len(df[4].unique())), df[4].unique())
plt.xlabel('Types')
plt.ylabel('Frequency')
plt.title('Histogram for class')
plt.show()
```

```python
#plot scatter plots
#there is nothing special correlation here.

plt.scatter(df[0], df[1], s=5)

plt.scatter(df[0], df[2], s=5)

plt.scatter(df[0], df[3], s=5)

plt.scatter(df[1], df[2], s=5)

plt.scatter(df[1], df[3], s=5)

plt.scatter(df[2], df[3], s=5)
```

```python
#Differentiation by each class

groups = df.groupby(4)
for group_name, group_data in grouped:
    print(f"Group: {group_name}")
    print(group_data)
    print()
```

```python
#Different histogram by each class
import pandas as pd
import matplotlib.pyplot as plt
for name, group in groups:
    plt.hist(group[0], bins=10, color='red') 
    plt.title(f'Histogram for sepal length in cm of {name}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
    
    plt.hist(group[1], bins=10) 
    plt.title(f'Histogram for sepal width in cm of {name}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
    
    plt.hist(group[2], bins=10, color = 'green')
    plt.title(f'Histogram for petal length in cm of {name}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
    
    plt.hist(group[3], bins=10, color = 'yellow')  
    plt.title(f'Histogram for petal width in cm of {name}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
```
